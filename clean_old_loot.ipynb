```python
import pandas as pd
import convert as c
```

```python
headers = ['date', 'player','item']
```

```python
snapshot = pd.read_csv('csv/2021_01_21_loot_snapshot.csv')
snapshot['date'] = snapshot['date'].map(c.dt)
```

```python
roster = pd.read_csv('csv/roster_with_class.csv')
roster
```

```python
snapshot['item'] = snapshot['item'].apply(lambda x: str(x).split('(')[0])
snapshot
```

```python
ss = snapshot.merge(roster, on='player', how='left').rename(columns={'item':'old_item_name'})
ss
```

```python
ss['player'] = ss['player'].map(c.deadshot)
```

```python
ss.sort_values(by='date',inplace=True)
```

Map old item names to new item names

```python
ss['item'] = ss.apply(c.convert, axis=1)
```

Filter out items older than 1/12/2021

```python
cleaned_old_loot = ss[ss['date'] < pd.to_datetime('1/13/2021', format='%m/%d/%Y')]
```

```python
cleaned_old_loot = cleaned_old_loot[headers]
```

```python
cleaned_old_loot['item'] = cleaned_old_loot['item'].map(str.strip)
cleaned_old_loot[cleaned_old_loot['item'] == 'Ghoul Skin Tunic']
```

```python
cleaned_old_loot[headers].to_csv('csv/cleaned_old_loot.csv', index = False)
cleaned_old_loot[headers]
```
