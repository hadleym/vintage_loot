```python
import pandas as pd
```

```python
df = pd.read_csv('csv/2021_01_21_loot_snapshot.csv')
df = df.dropna()
```

```python
clazz = pd.read_csv('csv/roster_with_class.csv')
df2 = df.merge(clazz, on='player', how='left')
df2.rename(columns={'item':'old_item_name'},inplace=True)
```

```python
items = df2[['old_item_name','class']]
old_item_df = items.drop_duplicates()
```

```python
df_tier_3 = old_item_df.loc[old_item_df['old_item_name'].str.contains('T3')]
df_tier_3.reset_index(inplace=True)
df_tier_3 = df_tier_3.drop(['index'], axis=1)
```

```python
s1 = df_tier_3['old_item_name'].apply(lambda x: x.split('(')[0])
```

```python
df_tier_3.loc[:,'old_item_name'] = s1
```

```python
df_tier_3.sort_values('class',inplace=True)
df_tier_3.reset_index(inplace=True)
df_tier_3 = df_tier_3.drop(['index'],axis=1)
df_tier_3.to_csv('csv/old_item_names.csv', index=False)
df_tier_3
```
