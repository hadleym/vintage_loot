```python
headers = ['date', 'player', 'item']
```

Import and Merge Files

```python
df_loot1 = pd.read_csv('csv/2020_01_13_loot.csv')
df_loot1['date'] = df_loot1['date'].map(lambda x: '0' + x[:-2]+ '2021')
df_loot2 = pd.read_csv('csv/2020_01_14_loot.csv')
df_loot2['date'] = df_loot2['date'].map(lambda x: '0' + x[:-2]+ '2021')
df_loot3 = pd.read_csv('csv/2020_01_20_loot.csv')
df_loot4 = pd.read_csv('csv/2020_01_21_loot.csv')
df_loot4['date'] = df_loot4['date'].map(lambda x: '0' + x[:-2]+ '2021')
df_loot = df_loot1.append([df_loot2, df_loot3, df_loot4], ignore_index=True)
df_loot['player']= df_loot['player'].str.split('-',expand=True)[0]
df_loot['item'] = df_loot['item'].str.replace('[','').str.replace(']','')
df_loot['player'] = df_loot.loc[:,'player'].map(c.deadshot)
```

Map date to datetime

```python
df_loot['date'] = df_loot['date'].map(c.dt)
df_loot
```
